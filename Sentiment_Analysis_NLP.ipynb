```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
data = {
    'Review': [
        'This phone is excellent',
        'Battery life is amazing',
        'Very good camera quality',
        'I am happy with this purchase',
        'Excellent performance',
        'Fantastic product',
        'Very satisfied',
        'Highly recommended',
        'The product is terrible',
        'Battery drains quickly',
        'Very poor quality',
        'Waste of money',
        'Bad experience',
        'Not worth buying',
        'Terrible service',
        'Very disappointing'
    ],
    'Sentiment': [
        'Positive','Positive','Positive','Positive',
        'Positive','Positive','Positive','Positive',
        'Negative','Negative','Negative','Negative',
        'Negative','Negative','Negative','Negative'
    ]
}

df = pd.DataFrame(data)
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(df['Review'])

y = df['Sentiment']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=1,
    stratify=y
)
```

```python
model = MultinomialNB()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print(y_pred)
```

```python
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
import matplotlib.pyplot as plt

plt.bar(['Accuracy'], [accuracy])

plt.title('Sentiment Analysis Accuracy')
plt.ylabel('Score')

plt.show()
```

```python
review = ["This product is excellent"]

review_vector = vectorizer.transform(review)

prediction = model.predict(review_vector)

print("Sentiment:", prediction[0])
```

# Sentiment Analysis with NLP

## Objective
To classify customer reviews as Positive or Negative using Natural Language Processing techniques.

## Conclusion
A sentiment analysis model was developed using TF-IDF Vectorization and Multinomial Naive Bayes. The model successfully predicts the sentiment of customer reviews and achieved an accuracy of 60%.
